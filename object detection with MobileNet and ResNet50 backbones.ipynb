Download Results:

```python
import shutil
import os
from IPython.display import FileLink

def zip_folder(folder_path, output_filename):
    shutil.make_archive(output_filename, 'zip', folder_path)

# Example usage
folder_path = '/content/runs'  # Path to the folder you want to zip
output_filename = '/content/results'  # Output file path (without .zip extension)
zip_folder(folder_path, output_filename)

# Provide a link to download the file
output_filename_with_extension = output_filename + '.zip'
FileLink(output_filename_with_extension)
```
